```python
def fibonacci(n):
    fib_sequence = [0, 1]  
    
    for i in range(2, n):
        next_fib = fib_sequence[i - 1] + fib_sequence[i - 2]
        fib_sequence.append(next_fib)
    
    return fib_sequence

def main():
    n = 100
    fib_sequence = fibonacci(n)

    for index, value in enumerate(fib_sequence):
        print(f"Fibonacci[{index}] = {value}")

if __name__ == "__main__":
    main()
```
